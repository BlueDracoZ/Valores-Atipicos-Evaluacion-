```python


df["TC"].fillna(0,inplace = True)
print(df.isnull().sum())
```

TC se decidio asi por que todos los datos son 1 o 0 en su mayoria, hay muy pocos que se salen de este rango asi que se decidio asi por que si son nulos podrian ser simplemente definidos como 0 

```python
df['IVA'] = df['IVA'].fillna(df['IMPORTE'] * 0.16)
print(df.isnull().sum())
```

```python
df['IMPORTE'] = df['IMPORTE'].fillna(df['IVA'] * 6.25)
print(df.isnull().sum())
```

```python
df['IMPORTE'] = df['IMPORTE'].fillna(df['TOTAL MX'] * 0.84)
print(df.isnull().sum())
```

```python
df['IVA'] = df['IVA'].fillna(df['TOTAL MX'] * 0.16)
print(df.isnull().sum())
```

**Las variables de IMPORTE,IVA y TOTAL MXN se complementan y son dependientes la una de la otra asi que solo se utiliza el saber que el IVA actual mexicano es de 16% y se rellenan los datos**

```python
df["FOLIO"].fillna("N999999999",inplace = True)
print(df.isnull().sum())
```

```python
df["TIPO"].fillna("NO SE SABE",inplace = True)
print(df.isnull().sum())
```

Aqui se decidio poner eso ya que es un dato que la misma tienda define y se define como independiente por lo cual no tiene patrón o variable que lo defina o dependa de su resultado a simple vista
