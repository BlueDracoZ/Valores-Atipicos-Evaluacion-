```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Año2020.csv")
df
```

```python
df2=pd.read_csv("Año2023.csv")
df2
```

```python
dfsel=df[["TC","IMPORTE","IVA","TOTAL MX","TOTAL SAT"]]
```

```python
sns.boxplot(data=dfsel)
```

```python
print(dfsel.isnull().sum())
```

```python

y = dfsel
percentile25 = y.quantile(0.25)
percentile75 = y.quantile(0.75)
iqr = percentile75 - percentile25
lower_bound = percentile25 - (1.5 * iqr)
upper_bound = percentile75 + (1.5 * iqr)
cleaned_data = dfsel[(dfsel >= lower_bound) & (dfsel <= upper_bound)]

print(cleaned_data)
```

```python
print(cleaned_data.isnull().sum())
```

```python
x=cleaned_data["TC"].mean()
print(x)
cleaned_data["TC"].fillna(x,inplace = True)
```

```python
x=cleaned_data["IMPORTE"].mean()
print(x)
cleaned_data["IMPORTE"].fillna(x,inplace = True)
```

```python
x=cleaned_data["IVA"].mean()
print(x)
cleaned_data["IVA"].fillna(x,inplace = True)
```

```python
x=cleaned_data["TOTAL MX"].mean()
print(x)
cleaned_data["TOTAL MX"].fillna(x,inplace = True)
```

```python
x=cleaned_data["TOTAL SAT"].mean()
print(x)
cleaned_data["TOTAL SAT"].fillna(x,inplace = True)
print(cleaned_data.isnull().sum())
```

```python
sns.boxplot(data=cleaned_data)
```
